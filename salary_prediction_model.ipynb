step II import Dataset


```python
salary = pd.read_csv('https://raw.githubusercontent.com/YBIFoundation/Dataset/main/Salary%20Data.csv')
```

STEP III define target(y) and features

```python
y= salary['Salary']
x= salary[['Experience Years']]
```

```python
x.shape
```

```python
y.shape
```

STEP IV TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.75, random_state=2529)
x_train
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train
```

```python
# STEP V SELECT MODEL
from sklearn.linear_model import LinearRegression
```

```python
model= LinearRegression()
```

```python
#step VI train model(fit model)
model.fit(x_train, y_train)
```

```python
#intercept value
model.intercept_
```

```python
#slope value
model.coef_
```

```python
#step VII Prediction
y_pred=model.predict(x_test)
y_pred
```

```python
#step VIII
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```
