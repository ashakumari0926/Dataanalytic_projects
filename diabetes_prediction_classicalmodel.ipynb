```python
import pandas as pd
```

```python
diabetes= pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Diabetes.csv')
```

```python
diabetes.info()
```

```python
diabetes.head()
```

```python
diabetes.columns
```

```python
#step 3 independent variable(y) and features
y= diabetes['diabetes']
x= diabetes[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age']]
```

```python
x.shape
```

```python
y.shape
```

# New Section

```python
#step 4
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=2529)
x_train
```

```python
#step 5 select model
from sklearn.linear_model import LogisticRegression
model= LogisticRegression(max_iter=500)
```

```python
#step 6 train model
model.fit(x_train, y_train)
```

```python
# step7 predict
y_pred= model.predict(x_test)
y_pred
```

```python
#step8 accuracy score
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
